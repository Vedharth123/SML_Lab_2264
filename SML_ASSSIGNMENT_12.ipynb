```python
import pandas as pd
df=pd.read_csv('/content/breast_cancer_survival.csv')
df
```

```python
df['Gender']=df['Gender'].replace(['MALE','FEMALE'],[0,1])
df['Tumour_Stage']=df['Tumour_Stage'].replace(['I','II','III'],[1,2,3])
df['ER status']=df['ER status'].replace(['Positive','Negative'],[1,0])
df['PR status']=df['PR status'].replace(['Positive','Negative'],[1,0])
df['HER2 status']=df['HER2 status'].replace(['Positive','Negative'],[1,0])
df['Patient_Status']=df['Patient_Status'].replace(['Alive','Dead'],[1,0])
df['Histology']=df['Histology'].replace(['Infiltrating Ductal Carcinoma','Infiltrating Lobular Carcinoma','Mucinous Carcinoma'],[1,0,2])
df['Surgery_type']=df['Surgery_type'].replace(['Lumpectomy','Modified Radical Mastectomy','Simple Mastectomy','Other'],[0,1,2,3])
df
```

```python
df.fillna(0,inplace=True)
df
```

```python
y=df['Patient_Status']
y
```

```python
x=df.drop(['Patient_Status','Date_of_Surgery','Date_of_Last_Visit'],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

def diff_train_knn(x, y, test_sizes, random_state):
    # Indent the entire function body
    resu = {}
    for test_size in test_sizes:
        print(f"\nTest Size: {test_size}")
        resu[test_size] = {}
        x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=test_size, random_state=random_state)
        scaler = StandardScaler()
        x_train = scaler.fit_transform(x_train)
        x_test = scaler.transform(x_test)
        for n in range(1, 21):
            model = LogisticRegression()
            model.fit(x_train, y_train)
            y_pred = model.predict(x_test)
            accuracy = accuracy_score(y_test, y_pred)
            print(f"n = {n}, Accuracy = {accuracy:.4f}")
            resu[test_size][n] = accuracy
    return resu # Indent the return statement as well

test_sizes = [0.2,0.25, 0.3, 0.35]
resu = diff_train_knn(x, y, test_sizes, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

def diff_train_knn(x, y, test_sizes, random_state):
    # Indent the entire function body
    res = {}
    for test_size in test_sizes:
        print(f"\nTest Size: {test_size}")
        res[test_size] = {}
        x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=test_size, random_state=random_state)
        scaler = StandardScaler()
        x_train = scaler.fit_transform(x_train)
        x_test = scaler.transform(x_test)
        for n in range(1, 21):
            model = SVC()
            model.fit(x_train, y_train)
            y_pred = model.predict(x_test)
            accuracy = accuracy_score(y_test, y_pred)
            print(f"n = {n}, Accuracy = {accuracy:.4f}")
            res[test_size][n] = accuracy
    return res # Indent the return statement as well

test_sizes = [0.2,0.25, 0.3, 0.35]
res = diff_train_knn(x, y, test_sizes, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

def diff_train_knn(x, y, test_sizes, random_state):
    results = {}
    for test_size in test_sizes:
        print(f"\nTest Size: {test_size}")
        results[test_size] = {}
        x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=test_size, random_state=random_state)
        scaler = StandardScaler()
        x_train = scaler.fit_transform(x_train)
        x_test = scaler.transform(x_test)
        for n in range(1, 21):
            model = KNeighborsClassifier(n_neighbors=n)
            model.fit(x_train, y_train)
            y_pred = model.predict(x_test)
            accuracy = accuracy_score(y_test, y_pred)
            print(f"n = {n}, Accuracy = {accuracy:.4f}")
            results[test_size][n] = accuracy
    return results

test_sizes = [0.2,0.25, 0.3, 0.35]
re = diff_train_knn(x, y, test_sizes, random_state=42)
```

```python
import matplotlib.pyplot as plt
def plot_grid_bootstrapping(re):
    fig, ax = plt.subplots(figsize=(10, 6))
    for test_size in re:
        accuracies = list(re[test_size].values())
        n_neighbors = list(re[test_size].keys())
        ax.plot(n_neighbors, accuracies, label=f'Test Size: {test_size}')
    ax.set_xlabel('Number of Neighbors (n)')
    ax.set_ylabel('Accuracy')
    ax.set_title('Grid Bootstrapping for KNN')
    ax.legend()
    ax.grid(True)
    plt.show()
plot_grid_bootstrapping(re)
```

```python
def plot_grid_bootstrapping_svm(res):
    fig, ax = plt.subplots(figsize=(10, 6))
    for test_size in res:
        accuracies = list(res[test_size].values())
        n_neighbors = list(res[test_size].keys())
        ax.plot(n_neighbors, accuracies, label=f'Test Size: {test_size}')
    ax.set_xlabel('Iterations')
    ax.set_ylabel('Accuracy')
    ax.set_title('Grid Bootstrapping for SVM')
    ax.legend()
    ax.grid(True)
    plt.show()
plot_grid_bootstrapping_svm(res)
```

```python
def plot_grid_bootstrapping_logistic(resu):
    fig, ax = plt.subplots(figsize=(10, 6))
    for test_size in resu:
        accuracies = list(resu[test_size].values())
        iterations = list(resu[test_size].keys())
        ax.plot(iterations, accuracies, label=f'Test Size: {test_size}')
    ax.set_xlabel('Iterations')
    ax.set_ylabel('Accuracy')
    ax.set_title('Grid Bootstrapping for Logistic Regression')
    ax.legend()
    ax.grid(True)
    plt.show()
plot_grid_bootstrapping_logistic(resu)
```

```python
def plot_all_grid_bootstrapping(re, res, resu):
    fig, ax = plt.subplots(figsize=(10, 6))

    for test_size in re:
        accuracies_knn = list(re[test_size].values())
        n_neighbors = list(re[test_size].keys())
        ax.plot(n_neighbors, accuracies_knn, label=f'KNN - Test Size: {test_size}')

    for test_size in res:
        accuracies_svm = list(res[test_size].values())
        iterations_svm = list(res[test_size].keys())
        ax.plot(iterations_svm, accuracies_svm, label=f'SVM - Test Size: {test_size}', linestyle='--')

    for test_size in resu:
        accuracies_logistic = list(resu[test_size].values())
        iterations_logistic = list(resu[test_size].keys())
        ax.plot(iterations_logistic, accuracies_logistic, label=f'Logistic - Test Size: {test_size}', linestyle=':')

    ax.set_xlabel('Iterations/Number of Neighbors')
    ax.set_ylabel('Accuracy')
    ax.set_title('Grid Bootstrapping for All Models')
    ax.legend()
    ax.grid(True)
    plt.show()

plot_all_grid_bootstrapping(re, res, resu)
```
