```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/Salary_Data.csv')
df
```

```python
y=df['Salary']
y
```

```python
x=df['YearsExperience']
x
```

```python
plt.scatter(x, y)
plt.title('YearsExperience vs Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
w0=1
w1=1
z=w0+w1*x
print(z)
yp=[]
for i in range(30):
  if z[i]>0:
    yp.append(1)
  else:
    yp.append(0)
```

```python
print(yp)
```

```python
error=y-yp
print(error)
```

```python
s=error**2
print(s)
```

```python
MSE=s.mean()
print(MSE)
```

```python
MAE=abs(error).mean()
print(MAE)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x1=x_train
y1=y_train
x2=x_test
y2=y_test
print(x2)
print(y2)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=20)
x1=x_train
y1=y_train
x3=x_test
y3=y_test
print(x3)
print(y3)
```
